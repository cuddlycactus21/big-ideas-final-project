# Luke's Dataset
https://data.wprdc.org/dataset/parks/resource/fa329e3d-89ff-4708-8ddf-81bfedcad11d

```python
import pandas as pd
parks = pd.read_csv("parks.csv")
parks.head(10)
```

Let's filter by parks only... Traffic Islands and Medians shouldn't count. Also, isolate only the important columns.

```python
parks = parks[parks.type == "Park"]
parks = parks[["name", "neighborhood"]]
parks
```

Now, let's see which neighborhood has the most parks.

```python
parks["neighborhood"].value_counts().plot(kind="barh", figsize=(6,10), title="Neighborhoods by Parks")
```

We can see that **South Side Slopes** and **Beechview** are tied for the best neighborhood with 6 parks each.
