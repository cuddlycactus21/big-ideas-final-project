# Luke's Dataset
https://data.wprdc.org/dataset/parks/resource/fa329e3d-89ff-4708-8ddf-81bfedcad11d

```python
import pandas as pd
parks = pd.read_csv("parks.csv")
parks.head(10)
```

Let's filter by parks only... Traffic Islands and Medians shouldn't count.

```python
parks = parks[parks.type == "Park"]
parks
```

Now, let's see which neighborhood has the most parks.

```python
parks["neighborhood"].value_counts().plot(kind="barh", figsize=(6,10))
```
